```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("this my x data")
plt.ylabel("this is my y data")
plt.title("line plot chart")
plt.grid()
plt.show()
```
