```python
import numpy as np
```

```python
#1.linear interpolation
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import matplotlib. pyplot as plt
```

```python
##interpolate the data using linear interpolation 
```

```python
plt.scatter(x,y)  # to add more points into this data we  use linear interpoplation 
```

```python
x_new = np.linspace(1,5,10) # create new x values
y_interp = np.interp(x_new,x,y) #interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

```python
#2 cubic interpolation with scipy
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y,kind='cubic') # create a cubic interpolation
```

```python
x_new = np.linspace(1,5,30)
y_interp = f(x_new)
```

```python
plt.scatter(x_new,y_interp)
```

```python
#polynomial interpolation
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
plt.scatter(x,y)
```

```python
# interpolate the data using polynomial interpolation
p = np.polyfit(x,y,4) # fit a 2nd degree polynomial to the data
```

```python
x_new = np.linspace(1,5,40)
y_interp = np.polyval(p,x_new)
```

```python
plt.scatter(x_new,y_interp)
```
